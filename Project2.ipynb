#### For project 2, I am going to examine the sound signature and elapsed time of Top 40 hits to determine if these features are predictive of a song's current popularity on Spotify.

```python
import pandas as pd
from datetime import timedelta
import numpy as np

songs = pd.read_csv('C:\Users\Matt\SkyDrive\Documents\GA Data Science\Final Project\song_data.csv')
songs.date = pd.to_datetime(songs.date,format='%m/%d/%Y')
```

```python
songs.head()
```

#### To simplify analysis, find each song's top rank and the day it occured.

```python
unique = songs.uri.unique()
```

```python
top_rank = pd.DataFrame(columns=['uri','rank','date'])
```

```python
ranker = {}
for i in unique:
    ranker['uri'] = i
    unique_song = songs[songs.uri == i]
    min_rank = unique_song[unique_song['rank'] == unique_song['rank'].min()]
    min_rank = min_rank.reset_index()
    ranker['rank'] = min_rank['rank'][0]
    max_date = min_rank[min_rank['date']==min_rank['date'].max()]
    max_date = max_date.reset_index()
    ranker['date'] = max_date['date'][0]
    top_rank = top_rank.append(ranker,ignore_index=True)
```

```python
top_rank.head()
```

```python
songs2 = songs.copy()
songs2 = songs2.drop(['date','rank'],axis=1)
songs2 = songs2.drop_duplicates()
```

```python
top_songs = pd.merge(songs2,top_rank,on='uri')
```

```python
top_songs['time'] = top_songs.date.max() - top_songs.date
top_songs['days'] = (top_songs.time /np.timedelta64(1, 'D')).astype(int)
top_songs.head()
```

## The Features

#### The features in this model are the sound signature components that make up a given song, as well as the elapsed time (in days) since that song's top rank on the Billboard Top 40 list [Note: 'Current Day' is considered 9/27/2014, the last day data was collected for this model].  The sound signature components, including danceability, instrumentalness, and liveness, are the quantitative measures of a song in various categories.  Moreover, given the attrition of popularity over time, elapsed time is a strong feature to map the gradual fall of popularity.

```python
import matplotlib.pyplot as plt
features = ['danceability','duration','energy','instrumentalness','key','liveness','loudness','tempo','time_signature','days']

pd.scatter_matrix(top_songs[features],figsize=(20,20))
```

## Data Prep

#### Using an OLS linear regression model to map the continuous Y popularity variable.

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()
```

#### Setting my chosen features to value X

```python
X = top_songs[features]
X = X.values
```

#### Setting my target values to Y

```python
y = top_songs['popularity']
y = y.values
```

## Test / Train Split

#### Dividing the X and Y variables into a training set and a test set.  The training set will be used to train the model and the test set will be used to test the accuracy of the trained model

```python
from sklearn.cross_validation import train_test_split
xtrain,xtest,ytrain,ytest =  train_test_split(X, y)
```

```python
model = clf.fit(xtrain,ytrain)
```

### Train Score

```python
model.score(xtrain,ytrain)
```

#### The model demonstrates a relatively weak fit with an R-Squared of 0.33

### Test Score

```python
model.score(xtest,ytest)
```

#### The model performs nearly as well on the test set as it did on the training set with a nearly identical R-Squared value.

#### Examine the coefficients.

```python
pd.DataFrame(zip(features,model.coef_.T),columns=['Variable','Coefficient'])
```

#### P-Values 

```python
from sklearn import feature_selection
f = feature_selection.f_regression(X,y)
pd.DataFrame(zip(features,f[1].T),columns=['Variable','P-Value'])
```

#### Based on the coefficient and P-Value analysis, the variables 'Key' and 'Tempo' can be removed from the model as they have a low impact on the target popularity value and P-values greater than 0.05.  Surprisingly, time elapsed also has very little impact in this model and will also be removed.

#### New Features

```python
new_features = ['danceability','duration','energy','instrumentalness','liveness','loudness','time_signature']
```

#### Retrain model

```python
X = top_songs[new_features]
X = X.values
```

```python
y = top_songs['popularity']
y = y.values
```

```python
xtrain,xtest,ytrain,ytest =  train_test_split(X, y)
```

#### With less significant features removed, the model actually performs worse on the training set

#### Model performs slightly worse on the revised test set

```python
pd.DataFrame(zip(new_features,model.coef_.T),columns=['Variable','Coefficient'])
```

```python
pd.DataFrame(zip(new_features,f[1].T),columns=['Variable','P-Value'])
```
